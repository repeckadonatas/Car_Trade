```python
import pandas as pd
import numpy as np
from babel.numbers import format_currency
```

```python
df = pd.read_csv('csv/FINAL_SPINNY_900.csv')
```

```python
# Displaying all columns in file for simplicity

pd.set_option('display.max_columns', 20)
```

```python
# Removing 'Car_Name' and 'Engine_Type' columns
# Don't run this cell more than once!!! if needed, rerun the whole doc!

df.drop(['Car_Name', 'Engine_Type'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Changing values to uppercase

df[['Color','Body_Type', 'Fuel_Type']] = df[['Color','Body_Type', 'Fuel_Type']].apply(lambda x: x.str.title())
df.head()
```

```python
# Checking the count of repeating values in 'No_of_Owners' column

df['No_of_Owners'].value_counts()
```

```python
# Checking for duplicate rows.
# From the info() result there are 62 duplicate rows

df.duplicated().value_counts()
```

```python
# Returning only unique rows

df = df.drop_duplicates()
df
```

```python
# Adding an ID column

df['car_id'] = df.index + 1
df.head()
```

```python
# Checking 'Price' column before cleaning the data

df['Price']
```

```python
# Checking to see if there are values in different formats and with different characters

df['Price'].apply(type).value_counts()
```

```python
# Removing ',' characters from values in 'Price' column

df['Price'].replace(',', '', regex=True, inplace=True)
df
```

```python
# Checking 'Price' column after cleaning the data

df['Price']
```

```python
# Changing the data type

df['Price'] = df['Price'].astype(int)
```

```python
df['Price']
```

```python
# Formating the values in 'Price' column

df['Price'] = df['Price'].apply(lambda x: (x/100))
df
```

```python
# Checking to see if there are values in different formats and with different characters

df['Mileage_Run'].apply(type).value_counts()
```

```python
# Removing ',' characters from values in 'Mileage_Run' column

df['Mileage_Run'].replace(',', '', regex=True, inplace=True)
df
```

```python
# Changing the data type

df['Mileage_Run'] = df['Mileage_Run'].astype(int)
df
```

```python
# df['Mileage_Run'] = df['Mileage_Run'].map(lambda x:'{:,.0f}'.format(x).replace(',', ' '))
# df
```

```python
# Checking to see if there are values in different formats and with different characters

df['Mileage(kmpl)'].apply(type).value_counts()
```

```python
# Checking to see if the data types are the same in all values of 'Mileage(kmpl)' column

df['Mileage(kmpl)'].equals('Mileage(kmpl)')
```

```python
# Listing the values of 'Mileage(kmpl)' column

df['Mileage(kmpl)'].unique()
```

```python
# Cleaning the values

df['Mileage(kmpl)'].replace({'BS IV': '0.0', '105 bhp @ 4400 RPM': '105'}, regex=True, inplace=True)
df
```

```python
# Checking the result

df['Mileage(kmpl)'].unique()
```

```python
# Changing the data type of values in 'Mileage(kmpl)' column

df['Mileage(kmpl)'] = df['Mileage(kmpl)'].astype(float)
df
```

```python
# Listing the values of 'Model' column

df['Model'].unique()
```

```python
# Cleaning the values

df['Model'].replace({'New i20' : 'i20',
                             'Wagon R 1.0' : 'Wagon R', 
                             'Wagon R 1.2' : 'Wagon R',
                             'New Santro 1.1' : 'New Santro'}, regex=True, inplace=True)
df
```

```python
# Checking the result

df['Model'].unique()
```

```python
# Changing the column name

df.rename(columns = {'Mileage(kmpl)' : 'Mileage(l/100km)'}, inplace=True)
df
```

```python
# Listing the values of 'Mileage(l/100km)' column

df['Mileage(l/100km)'].unique()
```

```python
# Cleaning the values and checking the result

df['Mileage(l/100km)'].replace(0.0, np.nan, inplace=True)
df['Mileage(l/100km)'].unique()
```

```python
# Changing the values to conform to L/100km format
# Rounding the result to 2 decimal points

df['Mileage(l/100km)'] = df['Mileage(l/100km)'].map(lambda x: (100/x))
df['Mileage(l/100km)'] = df['Mileage(l/100km)'].round(2)
df
```

```python
# Listing the values of 'Fuel_Type' column

df['Fuel_Type'].unique()
```

```python
# Finding the rows where 'Fuel_Type' == 'Petrol+Cng'

df.loc[df['Fuel_Type'] == 'Petrol+Cng']
```

```python
# Listing the values of 'Make' column

df['Make'].unique()
```

```python
# Splitting the list of values in 'Make' column and returning them in a new row

df['Make'] = df['Make'].str.split(' ')
df = df.explode('Make')
df
```

```python
# Splitting the list of values in 'Fuel_Type' column and returning them in a new row

df['Fuel_Type'] = df['Fuel_Type'].str.split('+')
df = df.explode('Fuel_Type')
df
```

```python
# Checking the results

df['Fuel_Type'].unique()
```

```python
# Splitting the list of values in 'Model' column and returning them in a new row

df['Model'] = df['Model'].str.split(' ')
df = df.explode('Model')
df
```

```python
# Finding the rows with the same car_id

res = df[df.groupby('car_id')['Make'].transform('nunique') > 1]
res
```

```python
# Saving the cleaned data to a CSV file

df.to_csv(r'csv/car_data_1NF.csv', sep=',', encoding='utf-8', index=None, header='true')
```
